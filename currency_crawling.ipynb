```python
import requests
from bs4 import BeautifulSoup

# 네이버 금융의 환율 페이지 URL
url = "https://finance.naver.com/marketindex/exchangeList.naver"

# 페이지 요청
response = requests.get(url)

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

# summary가 "환전 고시 환율 리스트"인 테이블 선택
exchange_table = soup.select('td.tit')
temp = [name.text.strip() for name in exchange_table]
currency_name = temp[:4]
currency_name.append(temp[6])

exchange_table = soup.select('td.sale')
temp = [val.text.strip() for val in exchange_table]
currency_val = temp[:4]
currency_val.append(temp[6])

exchange_rates = {currency_name[i] : currency_val[i] for i in range(len(currency_name))}
print(exchange_rates)
```
