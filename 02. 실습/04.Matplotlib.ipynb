```python
import matplotlib.pyplot as plt

import matplotlib as mpl

# 한글 폰트 설정 (전역)
plt.rcParams['font.family'] = 'AppleGothic'
# 마이너스 폰트 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

# matplotlib

```python
# 샘플 데이터
data = [10, -20, 30, -40, 50]
labels = ['하나', '둘', '셋', '넷', '다섯']
# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.bar(labels, data, color='orange')
plt.title('한글 폰트 확인 및 마이너스 부호 테스트')
plt.xlabel('X축 레이블')
plt.ylabel('Y축 값')
# 그래프 출력
plt.show()
```

```python
plt.title('default style의 plot')
plt.plot(range(10, 41, 10), [x**2 for x in range(1,5)])
plt.show()
```

# 축 범위, tick 설정

tick : plot의 x축/y축 표시 지점
tick label : tick에 써진 숫자/글자

```python
# 축 범위 설정

plt.title('축 범위 설정')
plt.plot(range(10, 41, 10), [x**2 for x in range(1,5)])
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
# tick 라벨 설정

import numpy as np

x = np.arange(0, 4*np.pi, 0.1) 
c = np.cos(x)

plt.plot(x, c)
plt.title('x축과 y축의 tick 라벨 설정')
plt.xticks([np.pi*x for x in range(1, 5)], 'π 2π 3π 4π'.split())
plt.yticks([-1,0,1])
plt.show()
```

# 여러 개의 데이터 그리기

x, y데이터, 스타일 문자열을 반복하여 인수로 넘김

- 하나의 선을 그릴 때처럼 x 데이터나 스타일 문자열을 생락할 수 없음

```python
t = np.arange(0, 5, 0.5)

plt.title('여러 개의 선 그리기')

plt.plot(t, t, 'r-*', t, t**2, 'y--^')
plt.show()
```

```python
t = np.arange(0, 5, 0.5)
y = [t, t**2]
plt.title('여러 개')
style = ['b--o', 'r:s']
for y, style in zip(y, style):
    plt.plot(t, y, style)
plt.show()
```

```python
t = t = np.arange(0, 5, 0.5)
styles = [{'c':'b', 'ls':'--','marker':'o'},{'c':'r', 'ls':':','marker':'s'}]
y_data = [t, t**2]
for y, style in zip(y_data, styles) :
    plt.plot(t, y, **style)
plt.show()
```

# 범례

```python
plt.title('cos, sin 그래프')
plt.xlabel('time')
plt.ylabel('amplitude')
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.plot(x, c, ls='--', label = 'cosine')
plt.plot(x, s, ls=':', label = 'sine')
plt.legend(loc=0)
plt.show()
```

# Figure, Axes

- Figure은 프레임의 개념, 실제로 그림이 그려지는 곳은 axes
피규어 안에 axes 여러 개를 그릴 수도 있음

```python
# subplot

x = np.linspace(-3,3,20)
y_list = [x**n for n in range(1,5)]

figure, axes = plt.subplots(2,2)

plot_number = 0
for i in range(2):
    for j in range(2) :
        axes[i][j].plot(x, y_list[plot_number])
        axes[i][j].set_title(f'x^{plot_number+1}')
        plot_number += 1

plt.figure(figsize=(2,2))

plt.show()
```

```python
# subplot 축 공유

fig, ax0 = plt.subplots()   # 기본 subplot 생성
ax1 = ax0.twinx()           # 공유되는 x축을 갖는 두 번째 y축 설정
ax0.set_title("2개의 y축 한 figure에서 사용")  # 제목 설정
# 첫 번째 y축 (ax0)
ax0.plot([10, 5, 2, 9, 8], 'r-', label="y0")  # 빨간색 실선 그래프
ax0.set_ylabel("y0")                          # 첫 번째 y축 라벨 설정
# 두 번째 y축 (ax1)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")  # 초록색 점선 그래프
ax1.set_ylabel("y1")                                   # 두 번째 y축 라벨 설정
# x축 라벨 설정
ax0.set_xlabel("공유되는 x축")
plt.show()  # 그래프 출력
```

# 다양한 모양의 그래프

```python
# 막대그래프

x = 'A B C'.split()
y = [1, 2, 3]

fig, ax = plt.subplots(1,2)

plt.figure(figsize=(6,4))
ax[0].set_title('Bar')
ax[0].bar(x,y)

ax[1].set_title('Bar H')
ax[1].barh(x,y)

plt.show()
```
