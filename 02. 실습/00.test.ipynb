```python
# Parameters, Arguments는 혼용되어 이야기하지만 2개는 엄밀히 다른 개념입니다.
# ○ Parameters : 함수 정의부에 나오는 이름
# ○ Arguments : 함수를 호출할 때 실제로 전달하는 값
```

```python
# 위치 전용 인자(positional only parameter)는 오직 위치 인수(positional arguments)만 값의 전달이 가능하다. 
# 위치 전용 인자를 선언하는 방법은 슬래시(/)를 인자 값으로 넣고 좌측 부분에 위치 전용 인자를 선언하면 된다.
def posonly(posonly, /):
    print(posonly)
posonly("값을 입력해야 합니다.")

# 아래는 키워드 인자라 오류
posonly(posonly = "값을 입력해야 합니다.")
```

```python
# 키워드 전용 인자(keyword only parameter)도 기본값을 가질 수 있다. 
# 이때 기본값을 갖는 인자에 상응하는 인수(arguments)는 생략할 수 있다.
def keyonly(*, keyonly):
    print(keyonly)
keyonly(keyonly = "값을 입력해야 합니다.")


# 가변-위치 인자(var-positional parameters)는 명시된 인자 외에 추가적으로 위치 인수를 개수에 상관없이 유연하게 전달 받을 수 있다. 
# 가변 변수로 사용할 변수명 앞에 *를 앞에 하나 붙여 표기.
def var_positional(*args):
    print(type(args))
    print(args)
    return sum([_ for _ in args])

print(var_positional(1,2,3,4,5))

def var_keyword(**kargs):
    print(type(kargs))
    print(kargs)
    return kargs

print(var_keyword(key = "value", key2 = "value2"))

```

```python
def days(x, y):
    if y == 2:
        if x % 400 == 0:
            return 29
        else:
            return 28
    else:
        # y가 7 이하면 홀수달이 31일, 8 이상이면 짝수달이 31일
        if (y % 2 != 0 and y < 7) or (y > 8 and y % 2 == 0):
            return 31
        else:
            return 30
     
############ 쌤 풀이
# 함수는 조건이 만족되면 즉시 종료된다.

# 윤년인지 확인하는 함수
def is_leap_year(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)
# 주어진 연도(year)와 월(month)을 입력받아 그 달의 일수를 반환
def days(year, month):
    # 각 월의 기본 일수
    month_days = {
        1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30,
        7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31
    }
    # 2월의 윤년 여부에 따라 일수를 조정
    if month == 2 and is_leap_year(year):
        return 29
    # 주어진 월의 일수 반환
    return month_days.get(month, "잘못된 월입니다.")
# 테스트 실행
print(days(1900, 11))  # 출력: 30
print(days(2004, 10))  # 출력: 31
print(days(1900, 2))   # 출력: 28
print(days(2000, 2))   # 출력: 29
print(days(2000, 13))   # 출력: 29        
```

```python
# 안쪽 함수에서 바깥쪽 함수에 선언된 변수 변경
# nonlocal 키워드

def level1():
    msg = "This is level 1."

    def level2():
        nonlocal msg
        msg = "Level2 is better than level1"

    level2()
    print(msg)

level1()
```

```python
# 첫 번째 Series 생성
s = pd.Series([9904312, 3448737, 2890451, 2466052],
              index=["서울", "부산", "인천", "대구"])
s.name = "인구"
s.index.name = "도시"

# 두 번째 Series 생성
s2 = pd.Series({"서울": 9631482, "부산": 3393191, "인천": 2632035, "대전": 1490158})


# 결측치는 NaN 
# ****** 결측치 처리 중요 : 보통 중위값이나 평균값으로 채움
# NaN 값은 float 자료형에서만 표현 가능하므로 다른 결과도 float 자료형으로 변환
print(s - s2)
```

```python
ds = s - s2

# 결측치만 추출
print("isnull :\n",ds.isnull())
# 결측치가 아닌 값들만 추출
print("notnull :\n", ds.notnull())

for k,v in ds.notnull().items():
    if v:
        print(k)

# 값이 True인 k,v 추출
ds[ds.notnull()]
```

```python
s = ds[ds.notnull()]

print(s.values.argmax()) # 최대값의 index 추출
print(s[[s.values.argmax()]]) # 해당 index 값 추출
```

```python
# tick : x축/y축 표시 지점
import numpy as np

x = np.arange(0, 4 * np.pi, 0.1) # 0부터 4π까지 0.1 간격으로 x값
c = np.cos(x) # 코사인 함수

plt.plot(x, c) # 그래프 그리기
plt.title("x축과 y축의 tick label 설정") # 제목 설정

# x축의 tick 값 설정
# π 표기
plt.xticks([np.pi, np.pi*2, np.pi*3, np.pi*4], [r'$\pi$', r'$2\pi$', r'$3\pi$', r'$4\pi$'])

# y축의 tick 값 설정
plt.yticks([-1,0,1])

plt.show() # 그래프 출력
```

```python
# 여러 개의 선 그리기
t = np.arange(0,5,0.2) # 0부터 5까지 0.2 간격으로 t값 생성
plt.title("하나의 plot에서 여러개의 선 그리기") # 제목 설정

# 여러 개의 선을 하나의 plot에 그리기
plt.plot(t, t, 'r--',
         t, t**2, 'bs:',
         t, t**3, 'g^-')

plt.show()
```

```python
t = np.arange(0,5,0.2) # 0부터 5까지 0.2 간격으로 t값 생성
plt.title("하나의 plot에서 여러개의 선 그리기") 

style = ['b--o', 'r:s']
y = [t, t**2] 

# zip 함수는 각각의 리스트를 짝 지어서 처리해줌
for y, style in zip(y, style):
    plt.plot(t,y,style)
plt.show()
```

```python
# 하나의 figure(프레임)에 여러개의 그래프를 그리려면 axes(캔버스)를 원하는 위치에 배치하기만 하면 된다.
# subflot(행의 수, 열의 수, 플롯번호)

x = np.linspace(-3,3,20) # -3에서 3까지 20개의 점 생성
y_list = [x, x**2] 

# 첫번재 figure
plt.figure(figsize=(3,2))
plt.plot(x, y_list[0])

# 첫번재 figure
plt.figure(figsize=(3,2))
plt.plot(x, y_list[1])

plt.show()
```

```python
x = np.linspace(-3,3,20) # -3에서 3까지 20개의 점 생성
y_list = [x, x**2, x**3, x**4] 

# 2x2 서브플롯 생성
figure, axes = plt.subplots(2,2)

# 서브플롯에 반복뭉르 사용해 그래프 그리기
plot_number = 0
for i in range(2):
    for j in range(2):
        axes[i][j].plot(x, y_list[plot_number])
        plot_number += 1 # 다음 y값으로 이동

plt.figure(figsize=(6,4))
plt.show()
```

```python
# 각 축을 공유하거나 분리하여 표시할 수 있다.
fig, ax0 = plt.subplots()   # 기본 subplot 생성
ax1 = ax0.twinx()           # 공유되는 x축을 갖는 두 번째 y축 설정
ax0.set_title("2개의 y축 한 figure에서 사용")  # 제목 설정
# 첫 번째 y축 (ax0)
ax0.plot([10, 5, 2, 9, 8], 'r-', label="y0")  # 빨간색 실선 그래프
ax0.set_ylabel("y0")                          # 첫 번째 y축 라벨 설정
# 두 번째 y축 (ax1)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")  # 초록색 점선 그래프
ax1.set_ylabel("y1")                                   # 두 번째 y축 라벨 설정
# x축 라벨 설정
ax0.set_xlabel("공유되는 x축")
plt.show() # 그래프 출력
```
