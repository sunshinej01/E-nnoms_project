```python
# í‚¤ì›Œë“œ ì „ìš© ì¸ì(keyword only parameter)ë„ ê¸°ë³¸ê°’ì„ ê°€ì§ˆ ìˆ˜ ìˆë‹¤. 
# ì´ë•Œ ê¸°ë³¸ê°’ì„ ê°–ëŠ” ì¸ìì— ìƒì‘í•˜ëŠ” ì¸ìˆ˜(arguments)ëŠ” ìƒëµí•  ìˆ˜ ìˆë‹¤.
def keyonly(*, keyonly):
    print(keyonly)
keyonly(keyonly = "ê°’ì„ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤.")


# ê°€ë³€-ìœ„ì¹˜ ì¸ì(var-positional parameters)ëŠ” ëª…ì‹œëœ ì¸ì ì™¸ì— ì¶”ê°€ì ìœ¼ë¡œ ìœ„ì¹˜ ì¸ìˆ˜ë¥¼ ê°œìˆ˜ì— ìƒê´€ì—†ì´ ìœ ì—°í•˜ê²Œ ì „ë‹¬ ë°›ì„ ìˆ˜ ìˆë‹¤. 
# ê°€ë³€ ë³€ìˆ˜ë¡œ ì‚¬ìš©í•  ë³€ìˆ˜ëª… ì•ì— *ë¥¼ ì•ì— í•˜ë‚˜ ë¶™ì—¬ í‘œê¸°.
def var_positional(*args):
    print(type(args))
    print(args)
    return sum([_ for _ in args])

print(var_positional(1,2,3,4,5))

def var_keyword(**kargs):
    print(type(kargs))
    print(kargs)
    return kargs

print(var_keyword(key = "value", key2 = "value2"))

```

```python
# í•˜ë‚˜ì˜ figure(í”„ë ˆì„)ì— ì—¬ëŸ¬ê°œì˜ ê·¸ë˜í”„ë¥¼ ê·¸ë¦¬ë ¤ë©´ axes(ìº”ë²„ìŠ¤)ë¥¼ ì›í•˜ëŠ” ìœ„ì¹˜ì— ë°°ì¹˜í•˜ê¸°ë§Œ í•˜ë©´ ëœë‹¤.
# subflot(í–‰ì˜ ìˆ˜, ì—´ì˜ ìˆ˜, í”Œë¡¯ë²ˆí˜¸)

x = np.linspace(-3,3,20) # -3ì—ì„œ 3ê¹Œì§€ 20ê°œì˜ ì  ìƒì„±
y_list = [x, x**2] 

# ì²«ë²ˆì¬ figure
plt.figure(figsize=(3,2))
plt.plot(x, y_list[0])

# ì²«ë²ˆì¬ figure
plt.figure(figsize=(3,2))
plt.plot(x, y_list[1])

plt.show()
```

```python
x = np.linspace(-3,3,20) # -3ì—ì„œ 3ê¹Œì§€ 20ê°œì˜ ì  ìƒì„±
y_list = [x, x**2, x**3, x**4] 

# 2x2 ì„œë¸Œí”Œë¡¯ ìƒì„±
figure, axes = plt.subplots(2,2)

# ì„œë¸Œí”Œë¡¯ì— ë°˜ë³µë­‰ë¥´ ì‚¬ìš©í•´ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plot_number = 0
for i in range(2):
    for j in range(2):
        axes[i][j].plot(x, y_list[plot_number])
        plot_number += 1 # ë‹¤ìŒ yê°’ìœ¼ë¡œ ì´ë™

plt.figure(figsize=(6,4))
plt.show()
```

```python
# ê° ì¶•ì„ ê³µìœ í•˜ê±°ë‚˜ ë¶„ë¦¬í•˜ì—¬ í‘œì‹œí•  ìˆ˜ ìˆë‹¤.
fig, ax0 = plt.subplots()   # ê¸°ë³¸ subplot ìƒì„±
ax1 = ax0.twinx()           # ê³µìœ ë˜ëŠ” xì¶•ì„ ê°–ëŠ” ë‘ ë²ˆì§¸ yì¶• ì„¤ì •
ax0.set_title("2ê°œì˜ yì¶• í•œ figureì—ì„œ ì‚¬ìš©")  # ì œëª© ì„¤ì •
# ì²« ë²ˆì§¸ yì¶• (ax0)
ax0.plot([10, 5, 2, 9, 8], 'r-', label="y0")  # ë¹¨ê°„ìƒ‰ ì‹¤ì„  ê·¸ë˜í”„
ax0.set_ylabel("y0")                          # ì²« ë²ˆì§¸ yì¶• ë¼ë²¨ ì„¤ì •
# ë‘ ë²ˆì§¸ yì¶• (ax1)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")  # ì´ˆë¡ìƒ‰ ì ì„  ê·¸ë˜í”„
ax1.set_ylabel("y1")                                   # ë‘ ë²ˆì§¸ yì¶• ë¼ë²¨ ì„¤ì •
# xì¶• ë¼ë²¨ ì„¤ì •
ax0.set_xlabel("ê³µìœ ë˜ëŠ” xì¶•")
plt.show() # ê·¸ë˜í”„ ì¶œë ¥
```

```python
# ë°” ì°¨íŠ¸(Bar chart)
y = [1,2,3]
x = ["A","B","C"]

fig, ax = plt.subplots(1,2) # 1í–‰ 2ì—´ì˜ ì„œë¸Œí”Œë¡¯ ìƒì„±

# ì²« ë²ˆì§¸ ì„œë¸Œí”Œë¡¯ : ì„¸ë¡œ ë§‰ëŒ€ ê·¸ë˜í”„
ax[0].set_title("Bar")
ax[0].bar(x,y)

# ë‘ ë²ˆì§¸ ì„œë¸Œí”Œë¡¯ : ê°€ë¡œ ë§‰ëŒ€ ê·¸ë˜í”„
ax[1].set_title("Bar H")
ax[1].barh(x,y)

plt.show()
```

```python
# íˆìŠ¤í† ê·¸ë¨(histogram)
# patches : ë§‰ëŒ€ í•˜ë‚˜í•˜ë‚˜ë¥¼ ê°€ë¦¬í‚´
# seed() : ëœë¤ ê°’ì´ ë™ì¼í•˜ê²Œ ë‚˜ì˜´
np.random.seed(42) # ëœë¤ ì‹œë“œ ê³ ì •
x = np.random.randn(1000) # ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ë‚œìˆ˜

plt.title("Histogram")
# hist() : íˆìŠ¤í† ê·¸ë¨ì„ ê·¸ë¦¬ê¸° ìœ„í•œ ë©”ì†Œë“œ
# bins : 
arrays, bins, patches = plt.hist(x, bins=10)
plt.setp(patches[2], edgecolor="r")
plt.show()
```

```python
# íˆìŠ¤í† ê·¸ë¨ - penguins dataset
penguins = sns.load_dataset("penguins")
penguins.head()
# xì¶• ì„¤ì •
sns.countplot(x=penguins["species"])
sns.histplot(data=penguins, y="flipper_length_mm", binwidth=5)
```

```python
# íˆìŠ¤í† ê·¸ë¨ - kde : ì»¤ë„ ë°€ë„, ë°ì´í„°ë¥¼ ì§ê´€ì ìœ¼ë¡œ ë³¼ ìˆ˜ ìˆìŒ
# ëˆˆê¸ˆ ê²½ê³„ê°€ ë¶ˆë¶„ëª…í•œ ê²ƒë“¤ì„ ë¶€ë“œëŸ½ê²Œ ë§Œë“¤ì–´ì¤Œ

sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
# multiple="stack" : ëˆ„ì ë˜ì–´ ìˆëŠ” í˜•íƒœ
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
# element="step" : ì„ ìœ¼ë¡œ í‘œí˜„
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", element="step")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", element="poly")
```
