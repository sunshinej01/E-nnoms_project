```python
import seaborn as sns
import pandas as pd
```

```python
# titanic, penguins
sns.set_theme(style = 'whitegrid')

titanic = sns.load_dataset('titanic')

penguins = sns.load_dataset('penguins')
penguins.head()
penguins.nunique()
penguins['species'].value_counts()
```

# seaborn

matplot을 기반으로 색상 테마와 통계용 차트 등의기능을 추가한 시각화 패키지

# 카운트플롯

범주형 자료, 이산분포

```python
# 카운트플롯


sns.countplot(data = titanic, x = 'class', hue = 'alive')
```

```python
# 가로 막대 그래프
sns.countplot(data = titanic, y = 'deck', hue = 'alive')
```

# 히스토그램


```python
# kde(분포곡선) 표시하기
# kernel density, 커널 함수를 겹치는 방법, 분포곡선 표시
sns.histplot(data=penguins, x = 'flipper_length_mm', kde = True)
```

# histplot() 표현법

```python
# hue 키워드 인수로 데이터 분리

sns.histplot(data=penguins, x = 'flipper_length_mm', hue="species")
```

```python
#multiple = 'stack'
sns.histplot(data=penguins, x = 'flipper_length_mm', hue="species", multiple='stack')
```

```python
# element = 'poly'
sns.histplot(data=penguins, x = 'flipper_length_mm', hue="species", element='poly')
```

# 히트맵 형태의 히스토그램 
- 상관계수를 나타낼 때 사용할 수 있음

```python
sns.histplot(penguins, x = 'bill_depth_mm', y = 'body_mass_g', hue='species')
```

```python
# 축의 값 중 하나가 이산된 값일 때

sns.histplot(penguins, x = 'bill_depth_mm', y = 'species', hue='species', legend=False)
```

# displot()
rug와 kde에 대해 동시에 표현 가능

```python
#기본 displot()

sns.displot(data=penguins, x = 'flipper_length_mm')
```

```python
#displot(kind = 'kde')
sns.displot(data=penguins, x = 'flipper_length_mm', kind='kde')
```

```python
#rug를 이용하여 두 가지 값의 kde 표현하기
sns.displot(data=penguins, x = 'flipper_length_mm', y = 'bill_length_mm', kind='kde', rug =True)
```

```python
# hue와 col을 이용하여 그래프 나누기

sns.displot(data=penguins, x = 'flipper_length_mm', hue = 'species', col='sex', kind='kde')
```

```python
#empirical cumulative distribution functions

sns.displot(data=penguins, y='flipper_length_mm', hue ='sex', col='species', kind='ecdf', height=4, aspect=.6)
```

```python
# axis label과 title 지정하기
g = sns.displot(data=penguins, y='flipper_length_mm', hue='sex', col='species', kind='kde', height=4, aspect=.7)

g.set_axis_labels('Density', 'Flipper length (mm)')
g.set_titles('{col_name} penguins')
```

# barplot()

```python
# 바플롯 만들기
sns.barplot(data=penguins, x = 'species', y='body_mass_g', hue='sex')
```

# boxplot()

```python
# 박스플롯 만들기
sns.boxplot(x=titanic['age'])
```

```python
# 박스플롯 여러 개로 만들기

sns.boxplot(data = titanic[['age','fare']], orient='h')
```

# violinplot()

```python
# 바이올린플롯
sns.violinplot(data=titanic, x = 'age', y = 'class')
```

```python
# 바이올린 플롯 스플릿
sns.violinplot(data=titanic, x = 'class', y='age', hue='alive', split=True)
```

# jointplot()

```python
sns.jointplot(data=penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
```

```python
# 단순선형 회귀 -> 데이터의 분포를 가장 잘 설명하는 선

adelie = penguins[penguins['species']=='Adelie']
sns.jointplot(data=adelie, x = 'bill_length_mm', y = 'bill_depth_mm', kind = 'reg')
```

# Heatmap()

```python


ttn_class_who = pd.pivot_table(titanic, 'survived', index='class', columns='who')
sns.heatmap(ttn_class_who, cmap='crest', annot= True, fmt = '.2f', linewidths=.5)
```
